```python

import json
import pandas as pd
df = pd.read_json('Processed_complete_dataset.json')
df.head()
```

```python
temp = df[df["taxonomy"] == "Temporal"]
snippetss = []

for index, claim in enumerate(temp.our_evidences):
    for i, evidence in enumerate(claim):
      for j, snippets in enumerate(evidence['search_results']):
          snippet = snippets["snippet"]
          snippetss.append([index,i, j, snippet])
import pandas as pd
temp_df = pd.DataFrame(snippetss, columns=['claim_index', 'question_index', 'snippet_index', 'snippet'])
temp_df
temp_df = temp_df[temp_df.claim_index>3511][:1000]
print(temp_df.head())
print(temp_df.shape)
from py_heideltime import heideltime

def process_chunk(text_chunk):
    return heideltime(text_chunk)
from tqdm import tqdm
tqdm.pandas()

temp_df["result"] = temp_df['snippet'].progress_apply(process_chunk)
```

```python
temp_df.to_csv("0_1000.csv")
```
